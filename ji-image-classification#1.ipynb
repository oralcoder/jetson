```python
import jetson.inference
import jetson.utils
from matplotlib import pyplot as plt
```

```python
net = jetson.inference.imageNet('googlenet')
```

```python
!wget raw.githubusercontent.com/oralcoder/jetson/master/images/polar_bear.jpg
```

```python
img = jetson.utils.loadImage('polar_bear.jpg')
```

```python
print(img.shape, type(img))
```

```python
print(img.size, img.width, img.height, img.channels)
```

```python
class_idx, confidence = net.Classify(img)
```

```python
class_desc = net.GetClassDesc(class_idx)
print(class_idx, class_desc, confidence)
```

```python
plt.figure(figsize=(16,16))
plt.imshow(img)
plt.show()
```
